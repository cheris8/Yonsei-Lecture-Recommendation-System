## KR-WordRank

```python
def KeywordExtractor(text):
    
    from krwordrank.hangle import normalize
    from krwordrank.word import KRWordRank
    
    text = normalize(str(text), english=True, number=True)
    
    wordrank_extractor = KRWordRank(
    min_count = 1, # 단어의 최소 출현 빈도수 (그래프 생성 시)
    max_length = 10, # 단어의 최대 길이
    verbose = True
    )

    keywords, rank, graph = wordrank_extractor.extract(text, beta=0.1, max_iter=10)
    
    return list(keywords.keys())
```

```python
data_['Keyword'] = data_['ReviewText'].apply(lambda x: KeywordExtractor(str(x).split('\n')))
```

```python
data_
```

```python
from krwordrank.hangle import normalize
from krwordrank.word import KRWordRank

corpus = list(data_['ReviewText'].values)
keywordList = list()

for doc in corpus:
    text = normalize(str(doc), english=True, number=True)
    
    wordrank_extractor = KRWordRank(
    min_count = 1,
    verbose = True
    )

    keywords, rank, graph = wordrank_extractor.extract(text, beta=0.1, max_iter=10)

    keywordList.append(list(keywords.keys()))

keywordList
```

### 코퍼스 단위

```python
texts = data_['ReviewText']
texts = [normalize(text, english=True, number=True) for text in texts]

wordrank_extractor = KRWordRank(
    min_count = 5, # 단어의 최소 출현 빈도수 (그래프 생성 시)
    max_length = 10, # 단어의 최대 길이
    verbose = True
    )

beta = 0.85 # PageRank의 decaying factor beta
max_iter = 10

keywords, rank, graph = wordrank_extractor.extract(texts, beta, max_iter)
```

```python
for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:30]:
    print('%8s:\t%.4f' % (word, r))
```

## soykeyword

```python
# 강의평 텍스트로부터 명사만 추출하여 새롭게 코퍼스를 구축합니다.

from konlpy.tag import Okt

corpus = list(data_['ReviewText'].values)
corpus_nouns = list()

okt = Okt()

for c in corpus:
    corpus_nouns.append(' '.join(okt.nouns(c)))
```

```python
# corpus-based extractor를 선언 및 훈련합니다.

from soykeyword.proportion import CorpusbasedKeywordExtractor

corpusbased_extractor = CorpusbasedKeywordExtractor(
    min_tf=20,
    min_df=2,
    tokenize=lambda x:x.strip().split(),
    verbose=True
)

corpusbased_extractor.train(corpus_nouns)
```

```python
### CorpusbasedKeywordExtractor().frequency()
# 각 단어가 코퍼스에서 몇 번 등장했는지 출현 빈도를 출력합니다.

word_list = ['쁠몰', '갓', '비율']
for word in word_list:
    print(word, corpusbased_extractor.frequency(word))
```

```python
### CorpusbasedKeywordExtractor().get_document_index()
# 해당 단어를 포함하고 있는 document idx를 반환합니다.

doc_idx = corpusbased_extractor.get_document_index('쁠몰')
doc_idx[:10]
```

```python
data_.iloc[doc_idx[105:110]]#['LectureName']
```

```python
### CorpusbasedKeywordExtractor().extract_from_word()
# 키워드 추출

keywords_from_word = corpusbased_extractor.extract_from_word(
    '쁠몰',
    min_score=0.8,
    min_frequency=100
)

keywords_from_word[:10]
```

```python
### CorpusbasedKeywordExtractor().extract_from_word()
# 키워드 추출

keywords_from_doc = corpusbased_extractor.extract_from_docs(
    doc_idx,
    min_score=0.8,
    min_frequency=100
)

keywords_from_doc[:10]
```

### matrix based

```python
# ttm을 생성합니다.

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=0.001)
x = vectorizer.fit_transform(corpus)

print(x.shape)
```

```python
# 정수인코딩을 수행합니다.

word2index = vectorizer.vocabulary_
index2word = sorted(
    vectorizer.vocabulary_,
    key=lambda x:vectorizer.vocabulary_[x]
)
```

```python
# matrix-based extractor를 선언 및 훈련합니다.

from soykeyword.proportion import MatrixbasedKeywordExtractor

matrixbased_extractor = MatrixbasedKeywordExtractor(
    min_tf=20,
    min_df=2,
    verbose=True
)

matrixbased_extractor.train(x, index2word)
```

```python
### MatrixbasedKeywordExtractor().extract_from_word()
# 키워드 추출

keywords_ = matrixbased_extractor.extract_from_word(
    '쁠몰',
    min_score=0.5,
    min_frequency=100)

keywords_[:10]
```
