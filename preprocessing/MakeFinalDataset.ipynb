## 최종 데이터셋 만들기

```python
import pandas as pd
```

```python
LectureTable = pd.read_csv('../data/2020LectureTable.csv', index_col=0)
LectureReview = pd.read_csv('../data/2020LectureReview.csv', index_col=0)
DepartmentInfo = pd.read_csv('../data/YonseiDepartmentInfo.csv', index_col=0)
```

```python
data = LectureTable.copy()
```

###  `LectureUrl` 컬럼 기준 중복 데이터 제거하기

```python
data.drop_duplicates('LectureUrl', keep='first', inplace=True)
```

### `LectureReview` 데이터셋의 컬럼들 추가하기

```python
LectureReview['ReviewText'].fillna('', inplace=True)
```

```python
# ReviewText 컬럼 생성
grouped = LectureReview.groupby('LectureUrl')
for key, group in grouped:
    res = ' '.join(LectureReview[LectureReview['LectureUrl']==key]['ReviewText'])
    data.loc[data['LectureUrl']==key, 'ReviewText'] = res

# ReviewText 컬럼 NA imputation
data['ReviewText'].fillna('', inplace=True)
```

### `DepartmentInfo` 데이터셋의 컬럼들 추가하기

```python
data['MajorId'] = data['LectureId'].apply(lambda x: x[:3])
```

```python
data = pd.merge(data, DepartmentInfo[['MajorId', 'MajorId_full', '모집단위', '총합']], how='left', on='MajorId')
```

```python
data.rename(columns={'모집단위':'MajorName', '총합':'MajorSize'}, inplace=True)
```

### 최종 데이터셋 저장하기
