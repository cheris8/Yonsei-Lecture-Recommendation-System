# Crawling

## 에브리타임에서 강의 기본 정보 및 상세 페이지 url 수집하기

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from bs4 import BeautifulSoup
from html_table_parser import parser_functions as parser

import time
import pandas as pd
```

```python
URL_20SPRING = 'https://yonsei.everytime.kr/timetable/2020/1'
URL_20FALL = 'https://yonsei.everytime.kr/timetable/2020/2'
USER_ID = '####'
USER_PW = '####'
```

### 2020-1 강의 목록 크롤링

```python
# 브라우저 생성 및 url 호출
driver = webdriver.Chrome('chromedriver')
driver.get(URL_20SPRING)
```

```python
# 로그인
driver.find_element_by_xpath('//*[@id="container"]/form/p[1]/input').send_keys(USER_ID)
driver.find_element_by_xpath('//*[@id="container"]/form/p[2]/input').send_keys(USER_PW)
driver.find_element_by_xpath('//*[@id="container"]/form/p[3]/input').send_keys(Keys.RETURN)
```

```python
time.sleep(5)
driver.find_element_by_xpath('//*[@id="sheet"]/ul/li[3]/a').click() # 팝업창 닫기
driver.find_element_by_xpath('//*[@id="container"]/ul/li[1]').click() # 수업 목록에서 검색 클릭
```

```python
# 웹페이지 최하단까지 스크롤
driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
```

```python
# html 파싱 및 table 태그 추출
page = driver.page_source
soup = BeautifulSoup(page, 'html.parser')
tables = soup.find_all('table')
```

```python
# 데이터프레임 생성
table = parser.make2d(tables[2])
data1 = pd.DataFrame(table[1:])
data1
```

```python
# 마지막 행 삭제
data1.drop(3651, axis=0, inplace=True)
data1
```

```python
# 강의 상세 페이지 url 및 강의평 (별점) 수집
class_urls = []
class_stars = []
for i in range(1, 3652):
    basexpath = '//*[@id="subjects"]/div[2]/table/tbody/tr['
    class_urls.append(driver.find_element_by_xpath(basexpath+str(i)+']/td[10]/a').get_attribute('href')) # 강의평 (url)
    class_stars.append(driver.find_element_by_xpath(basexpath+str(i)+']/td[10]/a').get_attribute('title')) # 강의평 (점수)
```

```python
print(len(class_urls))
print(len(class_stars))
```

```python
data1['class_urls'] = class_urls
data1['class_stars'] = class_stars
```

```python
data1
```

```python
driver.quit()
```
